```python
file_path = "icp.pdf"

reader = p4.LlamaMarkdownReader()
data = reader.load_data(file_path)
```

```python
text = ""

for i in range(len(data)):
    text += data[i].text_resource.text
```

```python
lines = text.split("\n")
```

```python
for i in range(len(lines)):
    lines[i] = lines[i].replace("##", "")
    lines[i] = lines[i].replace("**", "")
    lines[i] = lines[i].replace("\u200b", "-->")
    if (lines[i] != "" and lines[i][0] == "#"):
        lines[i] = lines[i][1::]
    lines[i] = lines[i].strip().rstrip()
```

```python
lines
```

```python
ret = []
for i in range(len(lines)):
    if (lines[i].lower() in ["references", "works cited"]):
        break
    if (lines[i] != ''):
        ret.append(lines[i])
lines = ret
```

```python
t = ""
for line in lines:
    t += line + "\n"
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def summarize(text, length):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer(Stemmer("english"))
    summarizer.stop_words = get_stop_words("english")
    summary = summarizer(parser.document, length)
    return summary

#def flatten(sentences):
#   if (sentences == []):
        
```
